## Exploring Brunei and Singapore GDP

```python
df_5 = df.copy()
```

```python
# Filter for rows of Brunei, Singapore, and East-Asia and Pacific countries

gdp_brn5 = df_5.loc[df_5['Country Code'] == 'BRN']
gdp_sgp5 = df_5.loc[df_5['Country Code'] == 'SGP']
gdp_eas5 = df_5.loc[df_5['Country Code'] == 'EAS']
```

```python
# Line graph of Brunei, Singapore, and East-Asia and Pacific countries GDP growth

plt.plot(gdp_brn5['Time'].values, gdp_brn5['GDP per capita (constant 2015 US$)'].values)
plt.plot(gdp_sgp5['Time'].values, gdp_sgp5['GDP per capita (constant 2015 US$)'].values)
plt.plot(gdp_eas5['Time'].values, gdp_eas5['GDP per capita (constant 2015 US$)'].values)

plt.xlabel('Year')
plt.ylabel('GDP per capital (constant 2015 US$)')

plt.title('GDP Per Capita')

plt.legend(('Brunei', 'Singapore', 'East-Asia and Pacific'))

#plt.axis([1960, 2022, 0, 40000])
plt.axis([1960, 2022, 0, 80000])
plt.grid()
```

```python
# Correlation between Singapore and Brunei GDP

X = gdp_brn5['GDP per capita (constant 2015 US$)']
Y = gdp_brn5['Time']
X2 = gdp_sgp5['GDP per capita (constant 2015 US$)']

plt.scatter(X,Y, label='Brunei')

plt.scatter(X2,Y, label='Singapore')

plt.legend()
```

There are little to no correlations between the GDP of both countries.


```python
# Correlation between Singapore and East-Asia and Pacific countries GDP

X = gdp_eas5['GDP per capita (constant 2015 US$)']
Y = gdp_sgp5['Time']

plt.scatter(X,Y, label='East-Asia & Pacific')

X2 = gdp_sgp5['GDP per capita (constant 2015 US$)']

plt.scatter(X2,Y, label='Singapore')

plt.legend()
```

While the development of Singapore's GDP may have been slighly correlated in the beginning, it has quickly outpaced all its surrounding countries with their rapid development and being a smaller country in comparison to their surrounding countries.

```python
# Correlation between Brunei and East-Asia and Pacific countries GDP

X = gdp_brn5['GDP per capita (constant 2015 US$)']
Y = gdp_brn5['Time']

plt.scatter(X,Y, label='Brunei')

X2 = gdp_eas5['GDP per capita (constant 2015 US$)']


plt.scatter(X2,Y, label='East-Asia & Pacific countries')

plt.legend()
```

While Brunei was founded a little later and with a higher GDP than the rest of the East-Asia & Pacific countries, their GDP has slowly been decreasing and becoming closer with other East-Asia & Pacific countries.
